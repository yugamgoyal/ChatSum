```python
import datetime
import pytz

# Function to filter messages from the past day
def filter_messages_past_day(messages):
    filtered_messages = []
    current_time = datetime.datetime.now(pytz.utc)
    one_day_ago = current_time - datetime.timedelta(days=1)

    for message in messages:
        timestamp = datetime.datetime.strptime(message["ts"], "%Y-%m-%dT%H:%M:%S.%fZ").replace(tzinfo=pytz.utc)
        if timestamp > one_day_ago and timestamp < current_time:
            filtered_messages.append(message)

    return filtered_messages
```

```python
import openai
from datetime import datetime, timedelta

# Define the time range for filtering messages
current_time = datetime.now()
past_day = current_time - timedelta(days=1)

# The text you want to summarize
text = ""  # Replace this with the text you want to summarize

# Open the text file
with open('chatHistory.txt', 'r') as file:
    # Read the contents of the file
    lines = file.readlines()
    # Iterate through each line and filter messages within the past day
    for line in lines:
        # Extract the timestamp from the line
        timestamp_str = line.split(']')[0].strip()
        timestamp = datetime.strptime(timestamp_str, '[%m/%d/%y, %I:%M:%S %p')
        # Check if the timestamp is within the past day
        if timestamp > past_day:
            # print(line.strip())  # Print the filtered message
            text += f"\n {line.strip()}"

print("Chat History:")
print(text)

# Set your OpenAI API key
openai.api_key = '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'

# Define the prompt to summarize the chat history
prompt = "Summarize the chat history from the past day, keep it short and simple, give bullet points! Be specific and keep details! Remember to not skip details, and use as much language from the text as possible instead of introducing more words!"

# Concatenate the prompt and the chat history
full_text = f"{prompt}\n{text}"

# Make an API call to the OpenAI GPT-3 model (gpt-3.5-turbo) for chat completion
response = openai.ChatCompletion.create(
    model="gpt-3.5-turbo",
    messages=[
        {"role": "system", "content": full_text},
        {"role": "user", "content": ""}
    ],
    max_tokens=60
)

# Extract the generated chat summary from the response
summary = response.choices[0].message.content.strip()

print("\nChat Summary:")
print(summary)
```
